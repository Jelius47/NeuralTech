```python
# 1. Load Your Data
import pandas as pd
import os
import json

# Load the JSON file
with open("/home/jelius/Downloads/chathistory.json", "r") as f:
    raw_data = json.load(f)

# Extract the list under "chathistory"
chat_data = raw_data["chathistory"]

# Create the DataFrame
df = pd.DataFrame(chat_data)

# Optional: Fill NA (cleaning)
df['message'] = df['message'].fillna("")
df['received_at'] = pd.to_datetime(df['received_at'])

#  2. Group by chat_id
grouped_chats = df.sort_values('received_at').groupby('chat_id')

# For huge datasets this leads to api limit crash

# #  3. Format each chat session
# def format_chat(chat_df):
#     lines = []
#     for _, row in chat_df.iterrows():
#         # Change this condition based on your actual 'responder' column, adjust if needed
#         sender = "Customer" if row.get('message_type') != 'chatbot' else "Bot"
#         timestamp = row['received_at']
#         message = row['message']
#         lines.append(f"{timestamp} - {sender}: {message}")
#     return "\n".join(lines)

# fix: limit to last fourty messages
def format_chat(chat_df, max_messages=40):
    chat_df = chat_df.tail(max_messages)
    lines = []
    for _, row in chat_df.iterrows():
        sender = "Customer" if row.get('message_type') != 'chatbot' else "Bot"
        timestamp = row['received_at']
        message = row['message']
        lines.append(f"{timestamp} - {sender}: {message}")
    return "\n".join(lines)

#  4. Set your OpenAI API key
from openai import OpenAI

client = OpenAI(
    # This is the default and can be omitted
    api_key=os.getenv("OPENAI_API_KEY")
,
)

def analyze_chat(chat_transcript):
    system_msg = {
        "role": "system",
        "content": "You are a helpful assistant that always responds with valid JSON output in the required format."
    }

    prompt = f"""
    Analyze the following chat and determine:
    1. If the customer is struggling.
    2. What are they struggling with?
    3. Overall sentiment (positive, neutral, negative).
    4. Summary of the conversation.

    Chat:
    {chat_transcript}

    Return a JSON with exactly these fields:
    {{
        "is_struggling": true or false,
        "struggle_reason": "...",
        "sentiment": "positive" or "neutral" or "negative",
        "summary": "..."
    }}
    """

    try:
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[system_msg, {"role": "user", "content": prompt}],
            temperature=0.3
        )

        # Extract the model's raw response
        content = response.choices[0].message.content.strip()

        # Check if the response starts with a valid JSON structure
        if not content.startswith("{"):
            print(" Unexpected response format:\n", content)
            raise ValueError("LLM did not return JSON.")

        # Safely parse JSON
        return json.loads(content)

    except Exception as e:
        print(f"Error analyzing chat: {e}")
        return {
            "is_struggling": None,
            "struggle_reason": f"LLM error: {str(e)}",
            "sentiment": "unknown",
            "summary": "Analysis failed."
        }

# 📊 6. Run analysis on each chat_id
results = []
for chat_id, chat_df in grouped_chats:
    transcript = format_chat(chat_df)
    analysis = analyze_chat(transcript)
    results.append({"chat_id": chat_id, **analysis})

# 📈 7. Convert results to DataFrame
results_df = pd.DataFrame(results)
results_df.head()
```
